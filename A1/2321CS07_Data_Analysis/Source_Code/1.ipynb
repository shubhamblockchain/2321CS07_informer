```python
import pandas as pd
data_etth1 = pd.read_csv(r'ETTh1.csv')
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV file
#df = pd.read_csv('your_large_file.csv')

# Convert the 'date' column to datetime
data_etth1['date'] = pd.to_datetime(data_etth1['date'], format='%d-%m-%Y %H:%M')

# Plot the line graph
plt.figure(figsize=(12, 6))
plt.plot(data_etth1['date'], data_etth1['OT'], color='b')
plt.xlabel('Date')
plt.ylabel('OT')
plt.title('Line Graph of OT Over Time')
plt.xticks(rotation=45)
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
data_electr = pd.read_csv(r'electricity.csv', error_bad_lines=False)
```

```python
data_electr.head()
```

```python
import pandas as pd

# Assuming your DataFrame is named 'data_electr'
# Define a function to clean and correct the date format
def clean_and_format_date(date_str):
    try:
        # Split the original date string into date and time parts
        date, time = date_str.split()

        # Replace any non-numeric characters in the date part
        date = ''.join(filter(str.isdigit, date))

        # Format the date and time parts to match '%Y-%m-%d %H:%M:%S'
        formatted_date = f"{date[:4]}-{date[4:6]}-{date[6:]} {time}"

        return formatted_date
    except:
        return None

# Apply the cleaning and formatting function to the 'date' column
data_electr['date'] = data_electr['date'].apply(clean_and_format_date)

# Convert the 'date' column to datetime
data_electr['date'] = pd.to_datetime(data_electr['date'], format='%Y-%m-%d %H:%M:%S')

# Now the 'date' column should contain correctly formatted datetime objects
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV file
#df = pd.read_csv('your_large_file.csv', delimiter='\t')

# Convert the 'date' column to datetime
data_electr['date'] = pd.to_datetime(data_electr['date'], format='%Y-%m-%d %H:%M:%S')

# Plot the line graph
plt.figure(figsize=(12, 6))
plt.plot(data_electr['date'], data_electr['OT'], color='b')
plt.xlabel('Date')
plt.ylabel('OT')
plt.title('Line Graph of OT Over Time')
plt.xticks(rotation=45)
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
